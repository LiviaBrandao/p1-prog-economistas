# Exerc√≠cio 2 ‚Äî Controle de vendas
<details>
<summary> Enunciado </summary>
‚Ä¢ Solicite ao usu√°rio o n√∫mero de vendas realizadas no dia.  

‚Ä¢ Para cada venda, solicite o nome do vendedor, valor da venda e meta individual.  

‚Ä¢ Calcule a percentagem atingida da meta e informe se o vendedor atingiu (‚â•100%) ou n√£o atingiu a meta.  

‚Ä¢ Utilize uma estrutura de repeti√ß√£o para processar todas as vendas.  

‚Ä¢ Ao final, exiba um resumo geral contendo:
- O total de vendas realizadas.  
- A m√©dia dos valores vendidos.
- Quantos vendedores atingiram a meta.
</details>

```python
from dataclasses import dataclass

@dataclass
class Venda:
    vendedor: str
    valor: float
    meta_individual: float

    def calcular_percentual_meta(self):
        return (self.valor * 100) / self.meta_individual

    def atingiu_meta(self):
        return self.calcular_percentual_meta() >= 100

vendas = []
total_atingiram_meta = 0

print('----- CONTROLE DE VENDAS-------')
total_vendas = int(input('Insira a quantidade de vendas: '))

for i in range(total_vendas):
    print(f'\n----- VENDA {i+1} -------')

    vendedor = input('Digite o nome do vendedor: ')
    valor = float(input('Digite o valor da venda: R$ '))
    meta = float(input('Digite a meta individual: R$ '))

    venda = Venda(vendedor, valor, meta)
    vendas.append(venda)

    percentual = venda.calcular_percentual_meta()
    if venda.atingiu_meta():
        total_atingiram_meta += 1

    print(f'{venda.vendedor} atingiu {percentual:.1f}% da meta de R${venda.meta_individual:.2f}')

# C√°lculos finais
total_valor_vendas = sum(venda.valor for venda in vendas)
media_vendas = total_valor_vendas / len(vendas) if vendas else 0

print(f'\n----- RESULTADOS -----')
print(f'O total em vendas realizadas foi de R$ {total_valor_vendas:.2f}')
print(f'A m√©dia dos valores vendidos foi de R$ {media_vendas:.2f}')
print(f'O total de vendedores que atingiram a meta foi de {total_atingiram_meta}')
```

# Exerc√≠cio 3 ‚Äî Fun√ß√µes e opera√ß√µes matem√°ticas
<details>
<summary> Enunciado </summary>
<pre>
Crie um programa que utilize fun√ß√µes para realizar opera√ß√µes diversas:
- Fun√ß√£o de Pot√™ncia:
  - Crie uma fun√ß√£o potencia(base, expoente) que retorne o valor de base elevado a expoente.<br>
-Fun√ß√£o de Fatorial:
  - Crie uma fun√ß√£o fatorial(n) que calcule o fatorial de n usando um la√ßo de repeti√ß√£o.<br>
- Fun√ß√£o de Convers√£o de Temperatura:
  - Crie uma fun√ß√£o converte_temp(valor, escala='C') que converta:
    - De Celsius para Fahrenheit, se escala='C'
    - De Fahrenheit para Celsius, se escala='F'<br>
O programa deve:
  - Solicitar ao usu√°rio os valores para testar cada fun√ß√£o.
  - Exibir os resultados de cada opera√ß√£o na tela com mensagens explicativas.
</pre>
</details>


```python
def rec_power(base, exp):
  if exp == 0:
    return 1
  else:
    return base * rec_power(base, exp - 1)

def rec_factorial(n):
    if n == 0:
        return 1
    else:
        return n * rec_factorial(n-1)

def celcius_to_fahrenheit(celcius):
    return (celcius * 9/5) + 32

def fahrenheit_to_celcius(fahrenheit):
    return (fahrenheit - 32) * 5/9

def show_menu():
    print('Digite a op√ß√£o desejada ' +
          '\n[1] - Calcular potencia de X ' +
          '\n[2] - Calcular fatorial de X ' +
          '\n[3] - Converter Celcius para Fahrenheit ' +
          '\n[4] - Converter Fahrenheit para Celcius ' +
          '\n[0] - Sair'+
          '\n\n')

while True:
    show_menu()
    option = input('\nSua op√ß√£o: ')

    match option:
      case "1":
          base = int(input('Digite a base: '))
          expoente = int(input('Digite o expoente: '))
          resultado = rec_power(base, expoente)
          print(f'O valor de {base} elevado a {expoente} √© {resultado}')
      case "2":
          n_factorial = int(input('Digite o n√∫mero para calcular o fatorial: '))
          print(f'O valor de {n_factorial} fatorial √© {rec_factorial(n_factorial)}')
      case "3":
          celcius = float(input('Digite uma temperatura em Celsius: '))
          print(f'A temperatura em Fahrenheit √© {celcius_to_fahrenheit(celcius):.2f}¬∞F')
      case "4":
          fahrenheit = float(input('Digite uma temperatura em Fahrenheit: '))
          print(f'A temperatura em Celsius √© {fahrenheit_to_celcius(fahrenheit):.2f}¬∞C')
      case "0":
          print('At√© logo! üëã')
          break
      case _:
          print('‚ùå Op√ß√£o inv√°lida! Tente novamente.')

```

# Exerc√≠cio 4 ‚Äî Manipula√ß√£o de arquivos
<details>
<summary> Enunciado </summary>
<pre>
- Crie um arquivo chamado vendas.txt contendo linhas no formato:
vendedor, produto, valor_venda
- O programa deve ler os dados do arquivo, calcular:
  - O total vendido por cada vendedor
  - O total geral de vendas
- Armazene as informa√ß√µes em uma lista de dicion√°rios.
- Gere um novo arquivo relatorio.txt com o formato:
Vendedor: total_vendido
- Exiba tamb√©m no terminal o vendedor com maior total de vendas.
</pre>
</details>
<details>
<summary> Arquivo vendas usado </summary>
<pre>
vendedor, produto, valor_venda
Alicia, tenis, 300
Cauan, casaco, 150
Alicia, jaqueta, 200
Alicia, calca jeans, 150
Cauan, saia, 90
Leticia, jaqueta, 400

</pre>
</details>

```python
from google.colab import files
from collections import defaultdict
import pandas as pd

uploaded = files.upload()

print('----- CADASTRO DE VENDAS BOUTIQUE MODA FASHION -------')
print('üêº Com pandas')
#using pandas
tfs = pd.read_csv('vendas.txt')
total_vendido = tfs.groupby('vendedor')[' valor_venda'].sum()
total_total_vendido = tfs[' valor_venda'].sum()
print(total_vendido)
print(total_total_vendido)

print('\nüìï Com dicionario')
#using dictionary
sales_bmf = tfs.to_dict('records')
sales_by_salesman = defaultdict(float)
total_in_sales = 0.0

for sale in sales_bmf:
    salesman = sale['vendedor']
    price = sale[' valor_venda']
    sales_by_salesman[salesman] += price
    total_in_sales += float(price)

top_salesman = max(sales_by_salesman, key=sales_by_salesman.get)
top_sale = sales_by_salesman[top_salesman]

with open('relatorio.txt', 'w', encoding='utf-8') as arquivo:
    for salesman, total in sales_by_salesman.items():
        line = f"{salesman}: {total}\n"
        arquivo.write(line)
print('TXT Gerado!')

print("üí∞ Vendas Por Vendedor:")
with open('relatorio.txt', 'r', encoding='utf-8') as arquivo:
    for line in arquivo:
        print(line.strip())

print("\nüèÜ Vendedor com mais Vendas:")
print(f"{top_salesman}: R${top_sale:.2f}")
print(f"\nüí∞ Total em vendas: \nR${total_in_sales:.2f}")

files.download('relatorio.txt')
```

# Exerc√≠cio 5 ‚Äî Visualiza√ß√£o de resultados em HTML
<details>
<summary> Enunciado </summary>
<pre>
- A partir do arquivo relatorio.txt gerado no exerc√≠cio anterior, crie uma p√°gina HTML
chamada relatorio_vendas.html.
- Essa p√°gina deve exibir:
  - Um t√≠tulo ‚ÄúRelat√≥rio de Vendas‚Äù
  - Uma tabela com duas colunas: Vendedor e Total Vendido
  - O vendedor com maior venda destacado em negrito e cor diferente.
- (Desafio opcional: se desejar, gere tamb√©m um gr√°fico de barras simples usando a biblioteca matplotlib e salve como imagem para ser exibida na p√°gina HTML.)

```python
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import base64
from io import BytesIO

with open('relatorio.txt', 'r') as f:
    data = f.read()

plt.figure(figsize=(9, 4))
cores = ['#ffeb3b' if v == top_salesman else '#3498db' for v in sales_by_salesman.keys()]
plt.bar(sales_by_salesman.keys(), sales_by_salesman.values(), color=cores)
plt.title('Vendas por Vendedor')
plt.xlabel('Vendedores')
plt.ylabel('Total Vendido (R$)')
plt.xticks(rotation=45)
plt.tight_layout()

# Salva o gr√°fico em mem√≥ria (BytesIO)
buffer = BytesIO()
plt.savefig(buffer, format='png', dpi=100, bbox_inches='tight')
buffer.seek(0)

# Converte para base64 para embedar no HTML
image_base64 = base64.b64encode(buffer.getvalue()).decode('utf-8')
plt.close()

html = f"""
<html>
<head><title>Relatorio de Vendas</title></head>
<body>
<h1>Relat√≥rio de Vendas</h1>
<table border="1">
<tr><th>Vendedor</th><th>Total Vendido</th></tr>
"""

for line in data.strip().split('\n'):
    if ':' in line:
        vendedor, total = line.split(': ')
        html += f"<tr><td>{vendedor}</td><td>R$ {total}</td></tr>"

html += f"""
</table>
<br>
<h2>Gr√°fico de Vendas</h2>
<img src="data:image/png;base64,{image_base64}" width="800">
</body>
</html>
"""

with open('relatorio_vendas.html', 'w') as f:
    f.write(html)

display(HTML(html))
files.download('relatorio_vendas.html')
```
